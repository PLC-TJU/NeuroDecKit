# 1.所有算法流程

```python
import sys

sys.path.append('E:\工作进展\小论文2023-RAVEplus\python程序\MI-All-dev')
from utils import extract_dict_keys

# 读取所有方法
DPA_METHODS = extract_dict_keys('transfer_learning.tl_classifier', 'TL_Classifier', 'check_dpa', 'prealignments')
FEE_METHODS = extract_dict_keys('transfer_learning.tl_classifier', 'TL_Classifier', 'check_fee', 'feature_extractions')
FES_METHODS = extract_dict_keys('transfer_learning.tl_classifier', 'TL_Classifier', 'check_fes', 'feature_selections')
CLF_METHODS = extract_dict_keys('transfer_learning.tl_classifier', 'TL_Classifier', 'check_clf', 'classifiers')
END_METHODS = extract_dict_keys('transfer_learning.tl_classifier', 'TL_Classifier', 'check_endest', 'estimators')
END_TO_END_METHODS = extract_dict_keys('transfer_learning.tl_classifier', 'TL_Classifier', 'check_endtoend', 'endtoends')

print('DPA_METHODS:', DPA_METHODS)
print('FEE_METHODS:', FEE_METHODS)
print('FES_METHODS:', FES_METHODS)
print('CLF_METHODS:', CLF_METHODS)
print('END_METHODS:', END_METHODS)
print('END_TO_END_METHODS:', END_TO_END_METHODS)
```

```python
import itertools

# 生成所有方法的组合
all_clf_list0 = list(itertools.product(
    range(len(DPA_METHODS)), 
    [1, 2, 5], 
    range(len(FES_METHODS)), 
    range(1, len(CLF_METHODS), 1), 
    [0], 
    [0]
))

all_clf_list1 = list(itertools.product(
    range(len(DPA_METHODS)), 
    [3, 4], 
    [0], 
    range(1, len(CLF_METHODS), 1), 
    [0], 
    [0]
))

all_clf_list2 = list(itertools.product(
    range(len(DPA_METHODS)), 
    [0], 
    [0], 
    [0], 
    range(1, len(END_METHODS), 1),
    [0]
))

all_clf_list3 = list(itertools.product(
    range(len(DPA_METHODS)), 
    [0], 
    [0], 
    [0], 
    [0], 
    range(1, len(END_TO_END_METHODS), 1)
))

all_clf_list = all_clf_list0 + all_clf_list1 + all_clf_list2 + all_clf_list3

# 将方法编号转换为方法名称并存入字典
algorithm_dict = {}
for indices in all_clf_list:
    methods = [
        DPA_METHODS[indices[0]], 
        FEE_METHODS[indices[1]], 
        FES_METHODS[indices[2]], 
        CLF_METHODS[indices[3]], 
        END_METHODS[indices[4]], 
        END_TO_END_METHODS[indices[5]]
    ]
    
    # 移除None值并构建关键字
    method_names = [method for method in methods if method != 'NONE']
    key = '-'.join(method_names)
    
    # 保存到字典中
    algorithm_dict[key] = list(indices)

# 打印字典中的部分内容以查看结果
for k, v in list(algorithm_dict.items())[:10]:
    print(f"{k}: {v}")

print(len(algorithm_dict))
```

# 1.2 保存所有算法流程及相应编号到csv文件

```python
import pandas as pd

# 创建DataFrame
data = []
for k, v in algorithm_dict.items():
    row = {
        'Algorithms': k,
        'Algorithm_ID': v,
        'DPA_METHODS': DPA_METHODS[v[0]],
        'FEE_METHODS': FEE_METHODS[v[1]] if v[1] != 0 else None,
        'FES_METHODS': FES_METHODS[v[2]] if v[2] != 0 else None,
        'CLF_METHODS': CLF_METHODS[v[3]] if v[3] != 0 else None,
        'END_METHODS': END_METHODS[v[4]] if v[4] != 0 else None,
        'END_TO_END_METHODS': END_TO_END_METHODS[v[5]] if v[5] != 0 else None,
        'DPA_METHODS_ID': v[0],
        'FEE_METHODS_ID': v[1],
        'FES_METHODS_ID': v[2],
        'CLF_METHODS_ID': v[3],
        'END_METHODS_ID': v[4],
        'END_TO_END_METHODS_ID': v[5]
    }
    data.append(row)

df = pd.DataFrame(data)

# 保存到CSV文件
df.to_csv('algorithms_all.csv', index=False)

# print(df.head())  # 打印前几行以查看结果

```

# 2. 部分选择的算法流程

```python
import itertools

# 定义所有方法
DPA_METHODS = ['TLDummy', 'EA', 'RA', 'RPA']
FEE_METHODS = [None, 'CSP', 'TRCSP', 'MDM', 'FGMDM', 'TS']
FES_METHODS = [None, 'ANOVA-K', 'ANOVA-P', 'MIC-K', 'MIC-P', 'PCA', 'LASSO', 'RFE', 'RFECV']
# CLF_METHODS = [None, 'SVM', 'LDA', 'LR', 'KNN', 'DTC', 'RFC', 'ETC', 'ABC', 'GBC', 'GNB', 'MLP', 'XGBoost', 'CatBoost', 'LightGBM']
CLF_METHODS = [None, 'SVM', 'LDA', 'LR', 'KNN', 'DTC', 'RFC', 'ETC', 'ABC', 'GBC', 'GNB', 'MLP']
END_METHODS = [None, 'RKNN', 'RKSVM', 'ABC-MDM', 'ABC-FGMDM', 'ABC-TSSVM', 'ABC-TSLDA', 'ABC-TSLR', 
               'MDWM', 'MEKT','MEKT-SVM','MEKT-LR']
END_TO_END_METHODS = [None, 'TRCA', 'DCPM', 'SBLEST']

# 挑选部分方法组合
dpa_list = [0, 1, 2, 3] # TLDummy, EA, RA, RPA
fee_list = [1, 5] # None, CSP, TS
fes_list = [0, 4, 6, 8] # None, MIC-P, LASSO, RFECV
clf_list = [1, 2, 3, 11] # None, SVM, LDA, LR, MLP
end_list = [5, 6, 7, 8, 9] # None, ABC-TSSVM, ABC-TSLDA, ABC-TSLR, MDWM, MEKT
end_to_end_list = [1, 2, 3] # None, TRCA, DCPM, SBLEST

# 生成挑选的部分方法组合
partial_clf_list1 = list(itertools.product(
    dpa_list, 
    fee_list, 
    fes_list, 
    clf_list, 
    [0], 
    [0]
))

partial_clf_list2 = list(itertools.product(
    dpa_list, 
    [0], 
    [0], 
    [0], 
    end_list, 
    [0]
))

partial_clf_list3 = list(itertools.product(
    dpa_list, 
    [0], 
    [0], 
    [0], 
    [0], 
    end_to_end_list
))

partial_clf_list = partial_clf_list1 + partial_clf_list2 + partial_clf_list3

# 将挑选的部分方法编号转换为方法名称并存入字典
partial_algorithm_dict = {}
for indices in partial_clf_list:
    methods = [
        DPA_METHODS[indices[0]], 
        FEE_METHODS[indices[1]], 
        FES_METHODS[indices[2]], 
        CLF_METHODS[indices[3]], 
        END_METHODS[indices[4]], 
        END_TO_END_METHODS[indices[5]]
    ]
    
    # 移除None值并构建关键字
    method_names = [method for method in methods if method]
    key = '-'.join(method_names)
    
    # 保存到字典中
    partial_algorithm_dict[key] = list(indices)

# 打印部分挑选的字典内容以查看结果
for k, v in list(partial_algorithm_dict.items())[:10]:
    print(f"{k}: {v}")

print(len(partial_algorithm_dict))
```

## 2.1 保存算法流程及相应编号信息到csv文件中

```python
import pandas as pd

# 创建DataFrame
data = []
for k, v in partial_algorithm_dict.items():
    row = {
        'Algorithms': k,
        'Algorithm_ID': v,
        'DPA_METHODS': DPA_METHODS[v[0]],
        'FEE_METHODS': FEE_METHODS[v[1]] if v[1] != 0 else None,
        'FES_METHODS': FES_METHODS[v[2]] if v[2] != 0 else None,
        'CLF_METHODS': CLF_METHODS[v[3]] if v[3] != 0 else None,
        'END_METHODS': END_METHODS[v[4]] if v[4] != 0 else None,
        'END_TO_END_METHODS': END_TO_END_METHODS[v[5]] if v[5] != 0 else None,
        'DPA_METHODS_ID': v[0],
        'FEE_METHODS_ID': v[1],
        'FES_METHODS_ID': v[2],
        'CLF_METHODS_ID': v[3],
        'END_METHODS_ID': v[4],
        'END_TO_END_METHODS_ID': v[5]
    }
    data.append(row)

df = pd.DataFrame(data)

# 保存到CSV文件
df.to_csv('algorithms_part.csv', index=False)

print(df.head())  # 打印前几行以查看结果

```
